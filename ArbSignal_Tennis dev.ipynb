## Winnaar

```python
## Note Kambi does noy yet scrape doubles

from rapidfuzz import process, fuzz
import pandas as pd

# Filter kambi_filtered_football based on criterion_label
filtered_kambi_winnaar = kambi_filtered_tennis[
    kambi_filtered_tennis['criterion_label'].str.contains('Wedstrijdnotering')
]

# Filter kambi_filtered_football based on criterion_label
filtered_toto_winnaar = toto_filtered_tennis[
    toto_filtered_tennis['Market Name'].str.contains('Wedstrijd')
]

# Preprocess strings: Remove extra whitespace and lowercase
def preprocess(text):
    return ' '.join(text.lower().split())
```

```python
import unicodedata

# Replace '-' with a space and remove accents in the 'names' column
filtered_kambi_winnaar['event_name'] = filtered_kambi_winnaar['event_name'].apply(
    lambda x: ''.join(
        char for char in unicodedata.normalize('NFKD', x.replace('-', ' '))
        if not unicodedata.combining(char)
    )
)
# Replace '-' with a space and remove accents in the 'names' column
filtered_toto_winnaar['Event Name'] = filtered_toto_winnaar['Event Name'].apply(
    lambda x: ''.join(
        char for char in unicodedata.normalize('NFKD', x.replace('-', ' '))
        if not unicodedata.combining(char)
    )
)

kambi_events = filtered_kambi_winnaar['event_name'].tolist()

# Function to find the best match for 'Event Name' in toto_filtered_football
def find_best_match(event_name):
    # Use a token-based similarity for better handling of substrings
    result = process.extractOne(event_name, kambi_events, scorer=fuzz.token_set_ratio, score_cutoff=90)
    if result is None:  # No match found
        return None
    match, score, _ = result
    return match

# Apply matching function to toto_filtered_football
filtered_toto_winnaar['matched_event'] = filtered_toto_winnaar['Event Name'].apply(find_best_match)
```

```python
# Merge the DataFrames using the standardized column
merged_df_winnaar = pd.merge(
    filtered_toto_winnaar,
    filtered_kambi_winnaar,
    left_on=['matched_event'], 
    right_on=['event_name'], 
    how='inner'  # Perform an inner join
)

merged_df_winnaar = merged_df_winnaar[(merged_df_winnaar['matched_event'].notnull())&(merged_df_winnaar['Outcome SubType'] != merged_df_winnaar['type'])]
```

```python
# Create 'OverUnderType'
toto_filtered_tennis_overunder['OverUnderType'] = toto_filtered_tennis_overunder['Market Name'].apply(
    lambda x: 'Sets' if 'Sets' in x and 'Games' not in x 
    else 'Games in Set' if 'Games' in x and 'Set' in x 
    else 'Games' if 'Games' in x 
    else 'Points' if 'Punten' in x 
    else None  # Default case if none of the conditions match
)

# Create 'OverUnderTime'
toto_filtered_tennis_overunder['OverUnderTime'] = toto_filtered_tennis_overunder['Market Name'].apply(determine_over_under_time)

# Create 'Team1' and 'Team2'
toto_filtered_tennis_overunder[['Team1', 'Team2']] = toto_filtered_tennis_overunder['Event Name'].str.split(' vs ', expand=True)

# Create 'OverUnderType2'
toto_filtered_tennis_overunder['OverUnderType2'] = toto_filtered_tennis_overunder.apply(
    lambda row: '1' if row['Team1'] in row['Market Name'] else (
        '2' if row['Team2'] in row['Market Name'] else 'Total team 1 and team 2'
    ),
    axis=1
)

# Create 'line'
toto_filtered_tennis_overunder['line'] = toto_filtered_tennis_overunder['Market Name'].apply(
    lambda x: float(x.split('Over/Under ')[-1]) if 'Over/Under ' in x else None
)
```

```python
# Perform the merge
merged_tennis_overunder = pd.merge(
    toto_filtered_tennis_overunder,
    kambi_filtered_tennis_overunder,
    left_on=['line', 'OverUnderType', 'OverUnderTime', 'OverUnderType2', 'matched_event'],
    right_on=['line', 'OverUnderType', 'OverUnderTime', 'OverUnderType2', 'event_name'],
    how='inner'  # Use 'inner' join to get only matching rows; adjust as needed
)
# Keep records with opposite outcomes
merged_tennis_overunder = merged_tennis_overunder[merged_tennis_overunder['outcome_english_label'] != merged_tennis_overunder['Outcome Name']]

# Display the merged DataFrame
merged_tennis_overunder.head()
```

## Match Winnaar & OverUnder & YesNo and do Arb Analysis

```python
# Perform the stacked union
total_tennis = pd.concat([merged_tennis_overunder, merged_df_winnaar, merged_tennis_yesno], ignore_index=True, sort=True)
total_tennis
```

```python
# Calculate Arbitrage Percentage
total_tennis['Arbitrage Percentage'] = (1 / total_tennis['Odds (Decimal)'] + 1 / total_tennis['odds']) * 100

# Identify Arbitrage Opportunities
total_tennis['Is Arbitrage'] = total_tennis['Arbitrage Percentage'] < 100

# Calculate Optimal Stakes if Arbitrage
total_stake = 1000
total_tennis['Stake A'] = total_tennis.apply(
    lambda row: (1 / row['Odds (Decimal)'] / (1 / row['Odds (Decimal)'] + 1 / row['odds']) * total_stake) if row['Is Arbitrage'] else 0,
    axis=1
)
total_tennis['Stake B'] = total_tennis.apply(
    lambda row: (1 / row['odds'] / (1 / row['Odds (Decimal)'] + 1 / row['odds']) * total_stake) if row['Is Arbitrage'] else 0,
    axis=1
)

total_tennis[['Event Name', 'Market Name', 'Outcome Name', 'outcome_label', 'Odds (Decimal)', 'odds', 'Arbitrage Percentage', 'Is Arbitrage', 'Stake A', 'Stake B']]    
```
