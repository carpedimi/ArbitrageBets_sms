```python
import requests
import numpy as np

# API endpoint with soccer as the sport key
url = "https://api.the-odds-api.com/v4/sports/soccer/odds/"

# Parameters for the API request
params = {
    "apiKey": "4917f0f8cf61057bf9aa75e0e032443b",  # Replace with your actual API key
    "regions": "eu",      # Region: Europe
    "oddsFormat": "decimal",  # Odds format: decimal
    "dateFormat": "iso",   # Date format: ISO 8601
    "markets": "h2h,spreads,totals"
}

# Send the request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse the JSON response
    print(data)  # Display the data
else:
    print(f"Request failed with status code {response.status_code}: {response.text}")
```

```python
import pandas as pd

# Example JSON response (assuming it is stored in a variable named 'response_data')
# response_data = <Your API JSON response here>

# Prepare a list to hold all rows of data
rows = []

# Loop through each event in the response
for event in data:
    event_id = event['id']
    sport_key = event['sport_key']
    commence_time = event['commence_time']
    home_team = event['home_team']
    away_team = event['away_team']
    
    # Loop through each bookmaker for the event
    for bookmaker in event['bookmakers']:
        bookmaker_key = bookmaker['key']
        bookmaker_title = bookmaker['title']
        last_update = bookmaker['last_update']
        
        # Loop through each market for the bookmaker
        for market in bookmaker['markets']:
            market_key = market['key']
            
            # Loop through each outcome in the market
            for outcome in market['outcomes']:
                outcome_name = outcome['name']
                price = outcome['price']
                point = outcome.get('point', None)  # Some outcomes have a 'point' key (for spreads)

                # Append data to rows list
                rows.append({
                    "Event ID": event_id,
                    "Sport Key": sport_key,
                    "Commence Time": commence_time,
                    "Home Team": home_team,
                    "Away Team": away_team,
                    "Bookmaker Key": bookmaker_key,
                    "Bookmaker Title": bookmaker_title,
                    "Last Update": last_update,
                    "Market Key": market_key,
                    "Outcome Name": outcome_name,
                    "Price": price,
                    "Point": point
                })

# Convert rows into a DataFrame
df = pd.DataFrame(rows)

# Display the DataFrame
print(df)
```
