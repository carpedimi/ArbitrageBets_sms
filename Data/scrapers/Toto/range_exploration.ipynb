```python
import csv
import ast
import glob

# Get all CSV files matching the pattern
csv_files = glob.glob("range_summary*.csv")  # Adjust the pattern if needed

# Initialize sets to store unique ranges
voetbal_ranges = set()
tennis_ranges = set()
basketbal_ranges = set()

# Process each CSV file
for csv_file in csv_files:
    try:
        with open(csv_file, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            header = next(reader, None)  # Skip header if it exists

            for row in reader:
                if len(row) < 2:  # Skip invalid rows
                    continue
                
                range_value = row[0]  # Range
                try:
                    sports_list = ast.literal_eval(row[1])  # Convert string to list
                except (SyntaxError, ValueError):
                    continue  # Skip malformed entries
                
                # Check for specific sports
                if "Voetbal" in sports_list:
                    voetbal_ranges.add(range_value)
                if "Tennis" in sports_list:
                    tennis_ranges.add(range_value)
                if "Basketbal" in sports_list:
                    basketbal_ranges.add(range_value)
    
    except Exception as e:
        print(f"Error reading {csv_file}: {e}")

# Convert sets to sorted lists for better readability
voetbal_ranges = sorted(voetbal_ranges)
tennis_ranges = sorted(tennis_ranges)
basketbal_ranges = sorted(basketbal_ranges)

# Output results
print("Unique Ranges where 'Voetbal' exists:", voetbal_ranges)
print("Unique Ranges where 'Tennis' exists:", tennis_ranges)
print("Unique Ranges where 'Basketbal' exists:", basketbal_ranges)
```

```python
def merge_ranges(ranges):
    # Convert string ranges to tuples of integers
    range_tuples = [tuple(map(int, r.split('-'))) for r in ranges]
    
    # Sort ranges by start value
    range_tuples.sort()

    merged = []
    start, end = range_tuples[0]

    for curr_start, curr_end in range_tuples[1:]:
        if curr_start <= end:  # Overlapping or continuous range
            end = max(end, curr_end)  # Extend the end boundary
        else:
            merged.append(f"{start}-{end}")  # Add previous range
            start, end = curr_start, curr_end  # Start new range
    
    # Add the last range
    merged.append(f"{start}-{end}")
    
    return merged

# Example input
voetbal_ranges = list(voetbal_ranges)

# Merge the ranges
merged_voetbal_ranges = merge_ranges(voetbal_ranges)

# Output the result
print(merged_voetbal_ranges)
```

```python
# Example input
tennis_ranges = list(tennis_ranges)

# Merge the ranges
merged_tennis_ranges = merge_ranges(tennis_ranges)

# Output the result
print(merged_tennis_ranges)
```
