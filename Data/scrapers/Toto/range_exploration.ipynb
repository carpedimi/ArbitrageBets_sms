```python
import csv
import ast

# Data (assuming the structure is in a CSV file)
csv_file = "range_summary.csv"  # Replace with the actual file name

# Initialize storage for sport occurrences
voetbal_ranges = []
tennis_ranges = []
basketbal_ranges = []

# Read the file
with open(csv_file, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header if it exists

    for row in reader:
        # Extract range, sports list
        range_value = row[0]
        sports_list = ast.literal_eval(row[1])  # Convert string representation of list to actual list

        # Check for specific sports
        if "Voetbal" in sports_list:
            voetbal_ranges.append(range_value)
        if "Tennis" in sports_list:
            tennis_ranges.append(range_value)
        if "Basketbal" in sports_list:
            basketbal_ranges.append(range_value)

# Output results
print("Ranges where 'Voetbal' exists:", voetbal_ranges)
print("Ranges where 'Tennis' exists:", tennis_ranges)
print("Ranges where 'Basketbal' exists:", basketbal_ranges)
```
